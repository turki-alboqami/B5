---

 1) Class vs Type

* **Class**: something you create (a blueprint).
* **Type**: what Python says a value is.

 A class is what you define, a type is what something *is*.

---

 2) Object vs Instance

* **Object**: any value in Python.
* **Instance**: an object created from a class.

 Every instance is an object.

---

 3) Variable vs Attribute

* **Variable**: a name that stores a value.
* **Attribute**: a variable that belongs to an object.

 Attribute = inside an object.

---

 4) Pure vs Impure Functions

* **Pure function**:

  * same input → same output
  * does not change anything outside

* **Impure function**:

  * may change outside data
  * can give different results

---

 5) f-string format specifiers

* `.2f` → 2 decimal places
* `10` → width
* `<` → left align
* `>` → right align
* `^` → center
* `,` → thousands separator
* `%` → percentage
* `x=` → show variable name and value
* `!r`, `!s`, `!a` → different ways to convert value to text

```python
def subtract_time(t1, t2):
    """Compute the difference between two times in seconds."""
    
    seconds1 = t1.hour * 3600 + t1.minute * 60 + t1.second
    seconds2 = t2.hour * 3600 + t2.minute * 60 + t2.second
    
    return seconds1 - seconds2
```

### Exercise

Write a function called `is_after` that takes two `Time` objects and returns `True` if the first time is later in the day than the second, and `False` otherwise.

Here's an outline of the function to get you started.

```python
def is_after(t1, t2):
    """Checks whether `t1` is after `t2`.

    >>> is_after(make_time(3, 2, 1), make_time(3, 2, 0))
    True
    >>> is_after(make_time(3, 2, 1), make_time(3, 2, 1))
    False
    >>> is_after(make_time(11, 12, 0), make_time(9, 40, 0))
    True
    """
    return None
```

```python
def is_after(t1, t2):
    if t1.hour > t2.hour:
        return True
    if t1.hour < t2.hour:
        return False
    
    if t1.minute > t2.minute:
        return True
    if t1.minute < t2.minute:
        return False
    
    return t1.second > t2.second
```

```python
is_after(make_time(3, 2, 1), make_time(3, 2, 0))
```

```python
is_after(make_time(3, 2, 1), make_time(3, 2, 1))
```
